```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
# filling null values of reviews by category vise. ex: if a clothing review is missing so it will fill the null value acc to the clothing category not footwear category
```

```python
df.isnull().sum()
```

```python
# making the column names into snake case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# Feature Engineering part
# creating a new column age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age','age_group']].head()
```

```python
# create column purchase_frequency_days bcoz in our data it is given weekly,fortnght etcso to improve readability we create a new column
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90,
    'Monthly':30
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# now we will check promo code applied and discount column
# while analysing we got to see that both the column contains same value i.e. Yes, but to make sure we will compare
(df['discount_applied'] == df['promo_code_used']).all()
#so now as both the columns are same we will drop one of the column
```

```python
df = df.drop('promo_code_used',axis=1)
```
